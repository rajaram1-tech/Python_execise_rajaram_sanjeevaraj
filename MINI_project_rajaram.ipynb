```python
import requests

from bs4 import BeautifulSoup

r = requests.get("https://www.flipkart.com/search?q=t-shirt%20under%20400&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off")
c = r.content
c
```

```python
soup=BeautifulSoup(c,"html.parser")
soup
```

```python
all=soup.find_all("div",{"class":"hCKiGj"})
all
```

```python
len(all)
```

```python

all[0]
```

```python
all[0].find_all("div",{"class":"syl9yP"})[0].text
```

```python
for item in all:   #to display all price value in the website
    print(item.find_all("div",{"class":"syl9yP"})[0].text)
    print(item.find_all("div",{"class":"Nx9bqj"})[0].text)
```

```python
l=[]
for item in all:   #to display all price value in the website
    d={}
    d["name"]=item.find_all("div",{"class":"syl9yP"})[0].text
   
    d["price"]=item.find_all("div",{"class":"Nx9bqj"})[0].text.replace("₹","Rs.")
    
    l.append(d)
```

```python
l
```

```python
import pandas
df=pandas.DataFrame(l)
```
